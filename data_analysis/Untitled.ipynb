```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parameters
mu = 0
sigma = 1

# X values
x = np.linspace(-10, 10, 1000)

# PDF
y = norm.pdf(x, mu, sigma)
c = norm.cdf(x,mu,sigma)

# Plot
plt.plot(x, c)
plt.title("Normal Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.show()
```
