```python
import numpy as np
```

```python

basis = np.array(
 [[3, 5],
 [1, 2]]
)
# declare vector v
v = np.array([2,3])
# create new vector
# by transforming v with dot product
new_v = basis.dot(v)
print(new_v) 
```

```python
# Declare i-hat and j-hat
i_hat = np.array([2, 0])
j_hat = np.array([0, 3])
# compose basis matrix using i-hat and j-hat
# also need to transpose rows into columns
basis = np.array([i_hat, j_hat]).transpose()
# declare vector v
v = np.array([1,1])
# create new vector
# by transforming v with dot product
new_v = basis.dot(v)
print(basis) # [2, 3]
```
