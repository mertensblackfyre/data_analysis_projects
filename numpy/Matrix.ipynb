```python
import numpy as np
import numpy.linalg
from sympy import *
```

```python
# Matrix Addition and Subtractions

A = np.array([2,1,3])
B= np.array([4,5,1])

add = A+B
sub = B-A

print(add)
print(sub)
```

```python
# Matrix Vector Multipication and Determint
ih = np.array([2,3])
jh = np.array([4,2])
vector = np.array([1,2])
matrix = np.array([ih,jh]).transpose()

answer = matrix.dot(vector)

det = np.linalg.det(matrix)
print(answer)
print(det)
```

```python
# Matrix Multipacation
# Transformation 1
i = np.array([2,1])
j = np.array([5,2])
t1 = np.array([i,j]).transpose()
# Transformation 2
i1 = np.array([8,4])
j1 = np.array([2,3])
t2 = np.array([i1,j1]).transpose()
# Multiply
combined = t1 @ t2
print(combined)
```
