```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rng = np.random.default_rng(seed=12345)
nsteps=10
draws = rng.integers(0, 2, size=nsteps)
```

```python
freq = 0
normal = []
for idx,n in enumerate(draws):
    if(n == 1):
        freq = freq + 1
        normal.append((freq/(idx+1), idx+1))
```

```python
x, y = zip(*normal)
plt.title("Simulations of tossing a coin N times")
plt.ylabel("Proportions of Head")
plt.xlabel("Number of tosses")
plt.plot(y, x)
plt.show()
```
